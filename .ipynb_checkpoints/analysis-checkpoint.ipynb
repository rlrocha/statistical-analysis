```python
import pandas as pd
import numpy as np
from outliers import smirnov_grubbs as grubbs
from scipy.stats import shapiro, levene, f_oneway
import matplotlib.pyplot as plt
```

```python
df_dados = pd.read_csv('dados/dados.csv')
df_dados.head()
```

```python
x_temp = np.arange(1, np.size(df_dados['C'].values)+1)

plt.plot(x_temp, df_dados['C'].values)
plt.ylabel('Resultado')
plt.xlabel('Experimento')
plt.title('Método C')
plt.grid()
plt.show()
```

```python
#metodos = ['A', 'B', 'C', 'D']
metodos = ['E', 'F', 'G']

#metodos = ['A', 'C']
#metodos = ['F', 'G']

for i in metodos:
    print('Método ' + i + ': %.2f%% (+/- %.2f%%)' % (np.mean(df_dados[i].values)*100, np.var(df_dados[i].values*100)))
```

```python
x_list = []

print('Teste de Grubb\n')
for i in metodos:
    if grubbs.min_test_indices(df_dados[i].values, alpha=.05) or grubbs.max_test_indices(df_dados[i].values, alpha=.05):
        print('Método ' + i + ': outliers')
        if grubbs.min_test_indices(df_dados[i].values, alpha=.05):
            indice = grubbs.min_test_indices(df_dados[i].values, alpha=.05)
            valor = grubbs.min_test_outliers(df_dados[i].values, alpha=.05)
            
        elif grubbs.max_test_indices(df_dados[i].values, alpha=.05):
            indice = grubbs.max_test_indices(df_dados[i].values, alpha=.05)
            valor = grubbs.max_test_outliers(df_dados[i].values, alpha=.05)
        
        print('Indíces:', indice)
        print('Valores: ', valor)
    
        x_list.append(grubbs.test(df_dados[i].values, alpha=.05))
    
    else:
        print('Método ' + i + ': sem outliers')
        x_list.append(df_dados[i].values)
```

```python
for i in range(len(x_list)):
    print('Método ' + metodos[i] + ': %.2f%% (+/- %.2f%%)' % (np.mean(x_list[i])*100, np.var(x_list[i]*100)))
```

```python
print('Teste de Shapiro - Wilk\n')

for i in range(len(x_list)):
    
    print('Método ' + metodos[i])
    
    stat, p = shapiro(x_list[i])
    print('Statistics=%f, p=%f' % (stat, p))

    alpha = 0.05
    if p > alpha:
        print('Distribuição parecer vir da distribuição Gaussiana (falha ao rejeitar H0)\n')
    else:
        print('Distribuição não parecer vir da distribuição Gaussiana (H0 rejeitada)\n')
```

```python
print('Teste de Levene\n')

alpha = 0.05

stat, p = levene(x_list[0], x_list[1], proportiontocut=alpha)
print('Statistics=%f, p=%f' % (stat, p))

if p > alpha:
    print('Não se rejeita a hipótese das distribuições terem a mesma variância (falha ao rejeitar H0)\n')
else:
    print('A hipótese das distribuições terem a mesma variância é rejeitada (H0 rejeitada)\n')
```

```python
print('Teste da Analise da Variância (ANOVA)\n')

stat, p = f_oneway(x_list[0], x_list[1])
print('Statistics=%f, p=%f' % (stat, p))

# interpret
alpha = 0.05
if p > alpha:
    print('Distribuições parecem ser iguais (falha ao rejeitar H0)')
else:
    print('Distribuições parecem ser diferentes (H0 rejeitada)')
```
